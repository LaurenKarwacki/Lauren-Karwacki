```python
# This code cell imports the necessary packages for the rest of the codeing steps.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This code cell imports the raw data found in my shared google drive folder into this notebook as a numpy array.
# The "delimiter = ','" allows the datapoints to be used, ignoring the commas that separate the datapoints in the raw data list.
url_hits = 'https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/acc_hits_hw5%20(1).csv'
acc_hits = np.loadtxt(url_hits, delimiter = ',')

url_fa = 'https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/acc_fa_hw5%20(1).csv'
acc_fa = np.loadtxt(url_fa, delimiter = ',')
```

```python
# This code cell prints the shape of each of the numpy arrays.
print(acc_hits.shape)
print(acc_fa.shape)
```

```python
# This code cell creates a list of 8001 equally spaced values between 0 and 4 seconds, which will correspond to each recorded data point.
time = np.linspace(0,4,8001)
time.shape
```

```python
# This code cell divides the length of each sample by the number of samples taken to calculate the duration of time between each sample taken.
samples = 8001
length = 4
sample_rate = length/samples
print(sample_rate)
```

```python
# This code creates the two subplots.
fig, ax = plt.subplots(2)

# This code plots the first trial of the hit data over time, labeling the x- and y-axes accordingly.
ax[0].plot(time, acc_hits[0])
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Voltage [$\mu V$]')

# This code plots the first trial of the false alarm data over time, labeling the x- and y-axes accordingly.
ax[1].plot(time, acc_fa[0])
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Voltage [$\mu V$]')
```

## Dominant Oscillation


* For the Hit Data graph, the dominant oscillation is about 1 Hz.
* For the False Alarm Data graph, the dominant oscillation is about 2-3 Hz.
* There is a slight difference between the first trials of acc_hits and acc_fa, with the frequency of the LFP data from the False Alarm dataset being slightly higher than the data from the Hits dataset.

```python
# This code calculates the ERP for the acc_hits dataset.
acc_hits.mean(0)

# This code plots the ERP data for the acc_hits dataset.
plt.plot(time, acc_hits.mean(0))

# This code plots a vertical line at 2 seconds.
plt.plot([2, 2], [-150,150], 'k', lw=2)

# This code plots a horizontal line at 0 microvolts.
plt.plot([0,4], [0,0], 'k', lw=2)

# This code labels the axes accordingly.
plt.xlabel('Time [s]')
plt.ylabel('Voltage [$\mu V$]')
```

## ERP Hits Analysis

* There is some activity after the vertical line, which indicates that there is brain activity in the ACC when the screen touch occurred.

```python
# This code calculates the ERP for the acc_fa dataset.
acc_fa.mean(0)

fig, ax = plt.subplots(2)

# This code plots the ERP data for the acc_fa dataset.
ax[0].plot(time, acc_fa.mean(0))

# This code plots a vertical line at 2 seconds.
ax[0].plot([2, 2], [-150,150], 'k', lw=2)

# This code plots a horizontal line at 0 microvolts.
ax[0].plot([0,4], [0,0], 'k', lw=2)

# This code labels the axes accordingly.
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Voltage [$\mu V$]')

# This code plots the ERP data for the acc_hits dataset and acc_fa dataset together.
ax[1].plot(time, acc_hits.mean(0))
ax[1].plot(time, acc_fa.mean(0))

# This code plots a vertical line at 2 seconds.
ax[1].plot([2, 2], [-150,150], 'k', lw=2)

# This code plots a horizontal line at 0 microvolts.
ax[1].plot([0,4], [0,0], 'k', lw=2)

# This code labels the axes accordingly.
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Voltage [$\mu V$]')
```

## ERP False Alarm Analysis

* There is no change in activity after the vertical line, which indicates that there is likely no brain activity in the ACC when the screen touch occurred.


* There is a difference between the ERPs for the hits (blue) and the false alarms (orange). The false alarms do not show brain activity when the screen touch occurred, whereas the hits did show brain activity in the ACC when the screen touch occurred.
