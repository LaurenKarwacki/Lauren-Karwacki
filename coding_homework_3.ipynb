```python
# This code cell imports the packages that will allow us to create numpy arrays (np), create data plots (plt), and perform statistical analyses (stats)
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# This code cell imports the raw data from my Pvalb and Sst datasets as numpy arrays. The url_(marker) are variables for each marker gene's urls for their raw data.
# The v_rest_(marker) is a variable representation of the numpy arrays for each marker's raw data.
url_pvalb = 'https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/v_rest_pvalb.csv'
v_rest_pvalb = np.loadtxt(url_pvalb)

url_sst = 'https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/v_rest_sst.csv'
v_rest_sst = np.loadtxt(url_sst)
```

```python
# This code cell assigns the variables mean_(marker) to the mean of each marker gene dataset.
mean_pvalb = np.mean(v_rest_pvalb)
print(mean_pvalb)

mean_sst = np.mean(v_rest_sst)
print(mean_sst)

# This code performs a t-test to determine if the means of the Pvalb and Sst datasets are statistically significantly different from each other.
print(stats.ttest_ind(v_rest_pvalb, v_rest_sst))

# This code tells python to create histograms displaying the frequency of each RMP's occurrance, but only if the p-value of the t-test is greater than 0.05.
pvalue = 7.159828660199036e-09
if pvalue > 0.05:
  fig, ax = plt.subplots(1,2)
  ax[0].hist(v_rest_sst)
  ax[1].hist(v_rest_pvalb)
  ax[0].set_xlabel('Mean Resting Membrane Potential (mV)')
  ax[0].set_ylabel('Frequency')
  ax[1].set_xlabel('Mean Resting Membrane Potential (mV)')
  ax[1].set_ylabel('Frequency')
```

```python
# This code cell creates a definition that runs the code above when "vrestHist()" is printed
def vrestHist():
  mean_pvalb = np.mean(v_rest_pvalb)
  print(mean_pvalb)

  mean_sst = np.mean(v_rest_sst)
  print(mean_sst)

  statistic, p_value_pvalb = stats.skewtest(v_rest_pvalb)
  print(p_value_pvalb)

  statistic, p_value_sst = stats.skewtest(v_rest_sst)
  print(p_value_sst)

  if pvalue > 0.05:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(v_rest_sst)
    ax[1].hist(v_rest_pvalb)
    ax[0].set_xlabel('Mean Resting Membrane Potential (mV)')
    ax[0].set_ylabel('Frequency')
    ax[1].set_xlabel('Mean Resting Membrane Potential (mV)')
    ax[1].set_ylabel('Frequency')

vrestHist()
```

```python
# This code cell prints the "vrestHist()" definition and prints the output of the code under the defintion.
vrestHist()
```

```python
# This code cell prints a scatterplot with Sst data on the x-axis and Pvalb data on the y-axis.
# The for loop looks at every point in the Sst and Pvalb datasets and assigns a color to them based on the combination of criteria of being > or < -60mV.
# The plt.xlabel creates the label "Sst Resting Membrane Potential (mV)" for the x-axis.
# The plt.ylabel creates the label "Pvalb Resting Membrane Potential (mV)" for the y-axis.
# The print(len(v_rest_(Marker))) prints the number of points in each dataset.
for i in range(len(v_rest_pvalb)):
  if v_rest_pvalb[i] > -60 and v_rest_sst[i] > -60:
    plt.scatter(v_rest_pvalb[i],v_rest_sst[i], color = 'red')
  elif v_rest_pvalb[i] > -60 and v_rest_sst[i] < -60:
    plt.scatter(v_rest_pvalb[i],v_rest_sst[i], color = 'green')
  elif v_rest_pvalb[i] < -60 and v_rest_sst[i] < -60:
    plt.scatter(v_rest_pvalb[i],v_rest_sst[i], color = 'blue')
  elif v_rest_pvalb[i] < -60 and v_rest_sst[i] > -60:
    plt.scatter(v_rest_pvalb[i],v_rest_sst[i], color = 'purple')
plt.xlabel('Sst Resting Membrane Potential (mV)')
plt.ylabel('Pvalb Resting Membrane Potential (mV)')

print(len(v_rest_sst))
print(len(v_rest_pvalb))
```
