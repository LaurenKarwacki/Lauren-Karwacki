```python
# This code cell imports the numpy and matplotlib.pyplot packages to be used later in the notebook.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This code cell imports all of the data to be used throughout the notebook as numpy arrays.
acc_hits = np.loadtxt('https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/acc_hits_hw6.csv', delimiter = ',')
acc_fa = np.loadtxt('https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/acc_fa_hw6.csv', delimiter = ',')
lc_hits = np.loadtxt('https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/lc_hits_hw6.csv', delimiter = ',')
lc_fa = np.loadtxt('https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/lc_fa_hw6.csv', delimiter = ',')
```

```python
# This code cell prints the shape of each dataset respectively.
print(acc_hits.shape)
print(acc_fa.shape)
print(lc_hits.shape)
print(lc_fa.shape)
```

```python
# This code cell creates a list of 8001 equally spaced values between 0 and 4 seconds, which will correspond to each recorded data point.
time = np.linspace(0,4,8001)
```

```python
# This code cell divides the length of each sample by the number of samples taken to calculate the duration of time between each sample taken.
sample_rate = 8001/4
print(sample_rate)
```

```python
# This code attributes "ax" to two separate subplots.
fig, ax = plt.subplots(2)

# This code plots the first trial of acc_hits in red and the first trial of lc_hits in blue on the first subplot, labeling the axes accordingly.
ax[0].plot(time, acc_hits[0], color = 'red')
ax[0].plot(time, lc_hits[0], color = 'blue')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Voltage ($\mu V$)')

# This code plots the first trial of acc_fa in red and the first trial of lc_fa in blue on the second subplot, labeling the axes accordingly.
ax[1].plot(time, acc_fa[0], color = 'red')
ax[1].plot(time, lc_fa[0], color = 'blue')
ax[1].set_xlabel('Time (s)')
ax[1].set_ylabel('Voltage ($\mu V$)')
```

# Dominant Oscillation of ACC and LC LFP Data

* **ACC Hits:** The dominant oscillation is about 2.5 Hz.
* **ACC False Alarms:** The dominant oscillation is about 1.5 Hz.
* **LC Hits:** The dominant oscillation is about 4 Hz.
* **LC False Alarms:** The dominant oscillation is about 2.5 Hz.

### General Observations:
* Though the voltages of the ACC and LC data are very different (LC data LFP data has a much smaller voltage), it appears that the hits data between the two brain regions are inversely proportional in terms of whether the voltage is positive or negative.
* Based on the graph, as the ACC hit data peaks positively, the LC hit data peaks negativley and vice versa. This also applies to false alarm data for both brain regions.

```python
# This code defines the time series for the acc_hits and lc_hits datasets, naming them "x" and "y" respectively.
x = acc_hits[0,:] - np.mean(acc_hits[0,:])
y = lc_hits[0,:] - np.mean(lc_hits[0,:])

# This code creates the variable "dt," which is the amount of time between each datapoint collected.
dt = time[1] - time[0]

# This code computes the Fourier transforms for the hits data.
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])

# This code compute the spectra for each Fournier transform and the combination of both for the hits data.
T = time[-1]
scale = 2 * sample_rate**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)
Syy = scale * (yf * yf.conj()).mean(0)
Sxy = scale * (xf * yf.conj()).mean(0)

# This code computes the coherence for the hits data.
cohr1 = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

# This code defines the frequency axis for the hits data.
f = np.fft.rfftfreq(8001, dt)

# This code creates the plot for the coherence over freqency for the hits data.
plt.plot(f, cohr1.real)

# This code defines the frequency range for the graph and scales the y-axis for coherence.
plt.xlim([0, 50])
plt.ylim([0, 1])

# This code labels the axes.
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
```

# Coherence for Hit Data

* The coherence is highest at the 1 Hz frequecy band.
* In general, there is an average decline of coherence as the frequency increases.

```python
# This code defines the time series for the acc_fa and lc_fa datasets, naming them "l" and "m" respectively.
l = acc_fa[0,:] - np.mean(acc_fa[0,:])
m = lc_fa[0,:] - np.mean(lc_fa[0,:])

# This code computes the Fourier transforms for the false alarm data.
lf = np.array([np.fft.rfft(l - l.mean()) for l in acc_fa])
mf = np.array([np.fft.rfft(m - m.mean()) for m in lc_fa])

# This code compute the spectra for each Fournier transform and the combination of both for the false alarm data.
T = time[-1]
scale = 2 * sample_rate**2 / T
Sll = scale * (lf * lf.conj()).mean(0)
Smm = scale * (mf * mf.conj()).mean(0)
Slm = scale * (lf * mf.conj()).mean(0)

# This code computes the coherence for the false alarm data.
cohr2 = np.abs(Slm) / (np.sqrt(Sll) * np.sqrt(Smm))

# This code defines the frequency axis for the false alarm data.
j = np.fft.rfftfreq(8001, dt)

# This code creates the plot for the corresponding dataset's coherence over freqency, with the hits data in red and the false alarm data in blue.
plt.plot(f, cohr1.real, color = 'red')
plt.plot(j, cohr2.real, color = 'blue')
plt.xlim([0, 50])
plt.ylim([0, 1])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Coherence')
```

# Coherence for Hit and False Alarm Data

* The coherence is highest at the 1 Hz frequecy band for the hits and false alarms, with the hits coherence being slightly higher than the false alarm coherence.
* In general, the coherence is stronger in the hits data throughout the entire frequency axis.
* Because there is more coherence for the hits data overall, the ACC and LC are communicating more during hits.
