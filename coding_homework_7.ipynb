```python
# This code cell imports all of the packages that will be used throughout the notebook.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# This code cell imports all of the individual raw datasets that will be used throughout the notebook.
spikes_left_trial1 = np.loadtxt('https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/spikes_left_trial1.csv', delimiter = ',')
spikes_left_trial2 = np.loadtxt('https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/spikes_left_trial2.csv', delimiter = ',')
spikes_right_trial1 = np.loadtxt('https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/spikes_right_trial1.csv', delimiter = ',')
spikes_right_trial2 = np.loadtxt('https://raw.githubusercontent.com/LaurenKarwacki/Lauren-Karwacki/main/spikes_right_trial2.csv', delimiter = ',')
```

```python
# This code cell prints the length of each array.
print(len(spikes_left_trial1))
print(len(spikes_left_trial2))
print(len(spikes_right_trial1))
print(len(spikes_right_trial2))
```

```python
# This code cell determines the number of spikes per second during the 30 second delay of the experiment by dividing the length of each array by the total time of delay.
fr_l1 = len(spikes_left_trial1)/30
print(fr_l1)

fr_l2 = len(spikes_left_trial2)/30
print(fr_l2)

fr_r1 = len(spikes_right_trial1)/30
print(fr_r1)

fr_r2 = len(spikes_right_trial2)/30
print(fr_r2)
```

# Spikes per Second

* Based on the spikes/sec data found, the neurons fire more before the first left turn trial (2.167 spikes/sec) compared to the second left turn trial (1.767 spikes/sec).
* The neurons fire less before the first right turn trial (1.5 spikes/sec) compared to the second right turn trial (2.3 spikes/sec).

```python
# This code indicates two subplots are going to be made.
fig, ax = plt.subplots(2)

# This code plots a line of spike data to display the left and right data from trial 1 on the first subplot.
# Each dataset is plotted on a separate level of the y-axis to be able to distinguish between the two datasets.
ax[0].plot(spikes_left_trial1, 1 * np.ones_like(spikes_left_trial1), '.')
ax[0].plot(spikes_right_trial1, 1.9 * np.ones_like(spikes_right_trial1), '.')

# This code sets the scales of the axes.
ax[0].set_xlim([0, 30])
ax[0].set_ylim([0, 3])

# This code labels the x-axis accordingly
ax[0].set_xlabel('Time [s]')

# This code creates a legend for each dataset.
ax[0].legend({'Right', 'Left'})

# This code removes the y-axis numerical ticks.
ax[0].set_yticks([])


# This code plots a line of spike data to display the left and right data from trial 2 on the second subplot.
# Each dataset is plotted on a separate level of the y-axis to be able to distinguish between the two datasets.
ax[1].plot(spikes_left_trial2, 1 * np.ones_like(spikes_left_trial2), '.')
ax[1].plot(spikes_right_trial2, 1.9 * np.ones_like(spikes_right_trial2), '.')

# This code sets the scales of the axes.
ax[1].set_xlim([0, 30])
ax[1].set_ylim([0, 3])

# This code labels the x-axis accordingly
ax[1].set_xlabel('Time [s]')

# This code creates a legend for each dataset.
ax[1].legend({'Right', 'Left'})

# This code removes the y-axis numerical ticks.
ax[1].set_yticks([])
```

## Spike Data Analysis

* The data from the left spikes appear to be more consistenly firing overall. The spikes appear to be spread over a greater distance.
* The data from the right spikes appear to be more bursted. The spikes have more space between "bursts" of them, especially seen in the right trial 2 data.

```python
# This code cell calculates the interspike intervals of each trial dataset, attributing them to their respective variables.
ISI_L1 = np.diff(spikes_left_trial1)
ISI_L2 = np.diff(spikes_left_trial2)
ISI_R1 = np.diff(spikes_right_trial1)
ISI_R2 = np.diff(spikes_right_trial2)
```

```python
# This code creates the four subplots that will be displayed.
fig, ax =  plt.subplots(2,2)

# This code sets the bin size for each histogram.
bins = np.arange(0, 2, 0.01)

# This code plots the histogram for the ISIs of the left trial 1 data in the top left subplot, labeling the axes and title accordingly.
ax[0,0].hist(ISI_L1, bins)
ax[0,0].set_xlabel('ISI [s]')
ax[0,0].set_ylabel('Counts')
ax[0,0].set_title('ISI from Left Trial 1')

# This code plots the histogram for the ISIs of the left trial 2 data in the bottom left subplot, labeling the axes and title  accordingly.
ax[1,0].hist(ISI_L2, bins)
ax[1,0].set_xlabel('ISI [s]')
ax[1,0].set_ylabel('Counts')
ax[1,0].set_title('ISI from Left Trial 2')

# This code plots the histogram for the ISIs of the right trial 1 data in the top right subplot, labeling the axes and title accordingly.
ax[0,1].hist(ISI_R1, bins)
ax[0,1].set_xlabel('ISI [s]')
ax[0,1].set_ylabel('Counts')
ax[0,1].set_title('ISI from Right Trial 1')

# This code plots the histogram for the ISIs of the right trial 2 data in the bottom right subplot, labeling the axes and title accordingly.
ax[1,1].hist(ISI_R2, bins)
ax[1,1].set_xlabel('ISI [s]')
ax[1,1].set_ylabel('Counts')
ax[1,1].set_title('ISI from Right Trial 2')
```

# Histogram Interpretation

* The neuron appears to fire more in bursts than consistently.
* If the neuron were to fire consistenly, it would gravitate toward one universal value. However, because there is not a lot of clustering around one value, it appears that there less consistent bursts of neuron firing.

```python
# This code attributes "ax" to two separate subplots.
fig, ax = plt.subplots(2)

# This code makes the time bins for every 1 second from 0-30 seconds.
fr_bins = np.arange(0,30,1)
# This code initializes the firing rate array.
fr_L1 = np.zeros(len(fr_bins)-1)
# This code creates a for loop that goes through each bin in the spikes_left_trial1 dataset.
for i in range(len(fr_bins)-1):
# This code finds the timestamp values that fall between each bin.
  spk_indexL1 = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])
# This code isolates the values that fit the previous line of code's criteria as the "true" value.
  true_valuesL1 = spikes_left_trial1[spk_indexL1]
# This code determines the firing rate of the spikes_left_trial1 datasets.
  fr_L1[i] = (len(true_valuesL1))/1

# This code does the same thing as the code above, but for the spikes_right_trial1 dataset.
fr_R1 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_indexR1 = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])
  true_valuesR1 = spikes_right_trial1[spk_indexR1]
  fr_R1[i] = (len(true_valuesR1))/1

# This code plots the left and right spike firing rate data for the first trials on the first subplot.
ax[0].plot(fr_bins[:-1], fr_L1, color = 'red')
ax[0].plot(fr_bins[:-1], fr_R1, color = 'blue')
ax[0].set_xlabel('Time [s]')
ax[0].set_ylabel('Firing Rate [Hz]')


# This code does the same thing as the code above, but for the spikes_left_trial2 dataset.
fr_L2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_indexL2 = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])
  true_valuesL2 = spikes_left_trial2[spk_indexL2]
  fr_L2[i] = (len(true_valuesL2))/1

# This code does the same thing as the code above, but for the spikes_right_trial2 dataset.
fr_R2 = np.zeros(len(fr_bins)-1)
for i in range(len(fr_bins)-1):
  spk_indexR2 = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])
  true_valuesR2 = spikes_right_trial2[spk_indexR2]
  fr_R2[i] = (len(true_valuesR2))/1

# This code plots the left and right spike firing rate data for the second trials on the second subplot.
ax[1].plot(fr_bins[:-1], fr_L2, color = 'red')
ax[1].plot(fr_bins[:-1], fr_R2, color = 'blue')
ax[1].set_xlabel('Time [s]')
ax[1].set_ylabel('Firing Rate [Hz]')
```

# Firing Rate Graph Interpretation

* In the first trials, it appears as though the left dataset has a higher firing rate overall when compared to the right dataset.
* In the second trials, it appears as though the right dataset has a higher firing rate overall when compared to the right dataset.
* In general, it appears as though the firing peaks are slightly off from each other. Sometimes the peaks line up, but not at the same magnitude, but sometimes there are opposing peaks, and sometimes both are not firing.
